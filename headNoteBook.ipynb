```python
maleDie = data[(data.sex == 1) & (data.DEATH_EVENT == 1)]
maleLive = data[(data.sex == 1) & (data.DEATH_EVENT == 0)]
femaleDie = data[(data.sex == 0) & (data.DEATH_EVENT == 1)]
femaleLive = data[(data.sex == 0) & (data.DEATH_EVENT == 0)]
mdl = len(maleDie.index)
mll = len(maleLive.index)
fdl = len(femaleDie.index)
fll = len(femaleLive.index)
total = mdl + mll + fdl + fll
percent = [mdl/ total, mll/ total, fdl/ total, fll/total]
labels = ["Male Deaths", "Male Survive", "Female Die", "Female Survive"]
plt.pie(percent, labels = labels, autopct='%1.1f%%')
plt.show()
```
